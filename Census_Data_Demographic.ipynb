```python
# Download ACS 2011-2015 5-year estimates for Oakland city, California on population size, median age, and median household income.
import censusdata
alldata = censusdata.download('acs5', 2019, censusdata.censusgeo([('state', '*'), ('county', '*')]), ["B01001_001E","B01001_002E","B01001_003E","B01001_004E","B01001_005E","B01001_006E","B01001_007E","B01001_008E","B01001_009E","B01001_010E",
"B01001_011E","B01001_012E","B01001_013E","B01001_014E","B01001_015E","B01001_016E","B01001_017E","B01001_018E","B01001_019E","B01001_020E",
"B01001_021E","B01001_022E","B01001_023E","B01001_024E","B01001_025E","B01001_026E","B01001_027E","B01001_028E","B01001_029E","B01001_030E",
"B01001_031E","B01001_032E","B01001_033E","B01001_034E","B01001_035E","B01001_036E","B01001_037E","B01001_038E","B01001_039E","B01001_040E",
"B01001_041E","B01001_042E","B01001_043E","B01001_044E","B01001_045E","B01001_046E","B01001_047E","B01001_048E","B01001_049E",
])

alldata.columns = [
"total_population","male_population","male_under5","male_5_9","male_10_14","male_15_17","male_18_19","male_20","male_21","male_22_24",
"male_25_29","male_30_34","male_35_39","male_40_44","male_45_49","male_50_54","male_55_59","male_60_61","male_62_64","male_65_66","male_67_69",
"male_70_74","male_75_79","male_80_84","male_85_and_over","female_population","female_under5","female_5_9","female_10_14","female_15_17","female_18_19","female_20","female_21","female_22_24",
"female_25_29","female_30_34","female_35_39","female_40_44","female_45_49","female_50_54","female_55_59","female_60_61","female_62_64","female_65_66","female_67_69",
"female_70_74","female_75_79","female_80_84","female_85_and_over"
]
alldata = alldata.reset_index()
alldata['state_fips'] = [alldata['index'][i].params()[0][1] for i in range(0,len(alldata))]
alldata['county_fips'] = [alldata['index'][i].params()[1][1] for i in range(0,len(alldata))]



demographic_data = alldata[['state_fips','county_fips','total_population','male_population','female_population']]
demographic_data['male_0_to_17'] = alldata['male_under5']+alldata['male_5_9']+alldata['male_10_14']+alldata['male_15_17']
demographic_data['male_18_to_49'] = alldata['male_18_19']+alldata['male_20']+alldata['male_21']+alldata['male_22_24']+alldata['male_25_29']+alldata['male_30_34']+alldata['male_35_39']+alldata['male_40_44']+alldata['male_45_49']
demographic_data['male_50_to_64'] = alldata['male_50_54']+alldata['male_55_59']+alldata['male_60_61']+alldata['male_62_64']+alldata['male_65_66']+alldata['male_67_69']
demographic_data['male_over_65'] = alldata['male_65_66']+alldata['male_67_69']+alldata['male_70_74']+alldata['male_75_79']+alldata['male_80_84']+alldata['male_85_and_over']

demographic_data['female_0_to_17'] = alldata['female_under5']+alldata['female_5_9']+alldata['female_10_14']+alldata['female_15_17']
demographic_data['female_18_to_49'] = alldata['female_18_19']+alldata['female_20']+alldata['female_21']+alldata['female_22_24']+alldata['female_25_29']+alldata['female_30_34']+alldata['female_35_39']+alldata['female_40_44']+alldata['female_45_49']
demographic_data['female_50_to_64'] = alldata['female_50_54']+alldata['female_55_59']+alldata['female_60_61']+alldata['female_62_64']+alldata['female_65_66']+alldata['female_67_69']
demographic_data['female_over_65'] = alldata['female_65_66']+alldata['female_67_69']+alldata['female_70_74']+alldata['female_75_79']+alldata['female_80_84']+alldata['female_85_and_over']


demographic_data['total_0_to_17'] = demographic_data['male_0_to_17']+demographic_data['female_0_to_17']
demographic_data['total_18_to_49'] = demographic_data['male_18_to_49']+demographic_data['female_18_to_49']
demographic_data['total_50_to_64'] = demographic_data['male_50_to_64']+demographic_data['female_50_to_64']
demographic_data['total_over_65'] = demographic_data['male_over_65']+demographic_data['female_over_65']

```
